```python
import pygame
import cv2
import time
import numpy as np
from tkinter import *
import cv2
```

```python
def face_eye():    
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
    cap = cv2.VideoCapture(0)
    font = cv2.FONT_HERSHEY_COMPLEX_SMALL
    while(True):
        # Capture frame-by-frame
        ret, video = cap.read()
        frame = cv2.flip(video,180)
        height,width = frame.shape[:2]
        # Our operations on the frame come here
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        #檢測人臉，並用vector保存人臉的坐標、大小（用矩形表示）
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
        global roi_gray
        global roi_color
        #臉部檢測
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = frame[y:y+h, x:x+w]
            #檢測臉部的眼睛，並用vector保存眼睛的坐標、大小
        
        eyes = eye_cascade.detectMultiScale(roi_gray)    
        #框出眼睛       
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)               

        t = 0 
        #沒偵測到眼睛                             
        while (not len(eyes)):
            
            ret, video = cap.read()
            frame = cv2.flip(video,180)
            height,width = frame.shape[:2]
            # Our operations on the frame come here
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            #檢測人臉，並用vector保存人臉的坐標、大小（用矩形表示）
            faces = face_cascade.detectMultiScale(gray, 1.3, 5)

            #臉部檢測
            for (x, y, w, h) in faces:
                cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
                roi_gray = gray[y:y+h, x:x+w]
                roi_color = frame[y:y+h, x:x+w]
                #檢測臉部的眼睛，並用vector保存眼睛的坐標、大小
        
            #time_start = time.time()
            time.sleep(0.1)
            t = t+0.1
            print(round(t,2))
            #time_end = time.time()
            #sum_t=(time_end - time_start)+sum_t
            
            eyes = eye_cascade.detectMultiScale(roi_gray)    
            #框出眼睛       
            for (ex, ey, ew, eh) in eyes:
                cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)

            if t >= 0.5:                        
                round_t = round(t,2)
                pygame.mixer.init()
                pygame.mixer.music.load('alert.mp3')
                pygame.mixer.music.play()
                cv2.putText(frame,'second:'+str(round_t),(100,height-20), font, 1,(255,255,255),1,cv2.LINE_AA)
                # frame外框畫一個紅色方框，線條寬度為 10   
                cv2.rectangle(frame, (0, 0), (width, height), (0, 0, 255), 10)                 
                print ("driver is sleeping !\n")
                result = "driver is sleeping !\n"  
                result_label.configure(text=result)
                time.sleep(0.05)
                break
        
        #按q關掉camera        
        cv2.imshow('frame', frame)
        # 每1毫秒監控一次鍵盤
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()
```

```python
root = Tk()
root.title("疲勞警示系統")
root.geometry('250x150')
root.config(cursor="arrow")
btn = Button(root, text="開啟!", command=face_eye , width = 20)
btn.pack(padx=10, pady=10)
result_label = Label(root,font = ('Arial', 11))
result_label.pack()
#video_loop()

root.mainloop()
```
